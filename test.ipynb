```python
x = np.array(fHz)
y = np.array(Uv)
# 进行样条曲线拟合
spline = make_interp_spline(x, y)

# 生成平滑曲线上的点
x_smooth = np.linspace(x.min(), x.max(), 100)
y_smooth = spline(x_smooth)

# 绘制拟合曲线
fig = plt.figure(figsize=(10,4))
plt.scatter(x, y,c='b', label='实际数据')
plt.plot(x_smooth, y_smooth,c='r', label='拟合曲线')

plt.legend()
plt.show()
```
